```python

class Estudiante:
    def __init__(self, nombre, apellido, cedula, semestre, materia):
        self.nombre = nombre
        self.apellido = apellido
        self.cedula = self.validar_cedula(cedula)
        self.semestre = self.validar_semestre(semestre)
        self.materia = materia

    def validar_cedula(self, cedula):
        if not cedula.isdigit():
            raise ValueError("La cédula solo debe contener números")
        if len(cedula) != 10:
            raise ValueError("La cédula debe tener exactamente 10 dígitos")
        return cedula

    def validar_semestre(self, semestre):
        if not isinstance(semestre, int):
            raise ValueError("El semestre debe ser un número entero")
        if semestre < 1 or semestre > 9:
            raise ValueError("El semestre debe estar entre 1 y 9")
        return semestre

    def mostrar_datos(self):
        print(f"Nombre: {self.nombre} {self.apellido}")
        print(f"Cédula: {self.cedula}")
        print(f"Semestre: {self.semestre}")
        print(f"Materia: {self.materia}")
```
